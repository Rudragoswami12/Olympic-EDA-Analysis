# DATA VISUALIZATION:-

# 1.Medal Count

```python

plt.hist(rb['Medal'])
plt.xlabel('Medal Count',fontsize=30,color='b')
plt.ylabel('Frequency',fontsize=30,color='c')
font = {'family':'serif','color':'red','size':50}
plt.title('Medal Count Distribution',fontdict=font)
plt.show()
```

# 2.Relationship between medal and team

```python
sns.scatterplot(x='Team', y='Medal', data=rb)
font = {'family':'serif','color':'green','size':30}
plt.title('Relationship Between Medal Count and Team',fontdict=font)

```

# 3.Sport Type:-

```python
df = plt.subplots(figsize = (10, 5))
mylabels = ["Athletics", "Hockey", "Ice Hockey", "Rowing", "Swimming"]
rb['Sport'].value_counts().head(5).plot(kind = 'pie',labels = mylabels, autopct="%1.1f%%")
plt.legend(title = "Sports",loc='right')
```

# 4.Age and Weight Analyasis:-

```python
sns.set(rc={'figure.figsize':(25,10)})
ax=sns.boxplot(x='Age', y ='Weight',data=rb)
ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
```

# 5.Visualization Season and There Gsmes:-

```python
plt.figure(figsize=(10,15))
sns.boxplot(x=rb['Season'], y=rb['Games'])
```
